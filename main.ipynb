```python
import cv2
import easyocr
import matplotlib.pyplot as plt
```

```python
# This needs to run only once to load the model into memory
reader = easyocr.Reader(['en'])
```

```python
# reading the image
img = cv2.imread('./sample/sample.png')

# run OCR
results = reader.readtext(img)
```

```python
plt.imshow(img)
```

```python
for res in results:
    print(res[1])
```
